```python
import pandas as pd

f_mcode_url = r"C:\Users\Drakharys\Desktop\JustLearning\MUSIC\subject-info_codes.csv"

f_codes = pd.read_csv(f_mcode_url, sep=';', encoding='latin1') # El archivo estña codificado como latin1, algun excel antiguo

f_codes.head()
```

```python
# Expando la variable a cada uno de los códigos para quitarme los NaN
f_codes['Variable'] = f_codes['Variable'].fillna(method='ffill')

f_codes.head(10)
```

```python
# Creo nuevas columnas a partir de la columna 'Codes', manejando valores NaN y usando directamente expand
f_codes[['Código', 'Descripción']] = (
    f_codes['Codes']
    .dropna() 
    .str.split(':', expand=True)  
)

# Elimino espacios en las nuevas columnas
f_codes['Código'] = f_codes['Código'].str.strip()
f_codes['Descripción'] = f_codes['Descripción'].str.strip('; ').str.strip()

print(f_codes.head(10))
```

```python
# Me cargo las columnas originales innecesarias
f_codes = f_codes.drop(columns=['Column', 'Notes', 'Codes'])

f_codes.head(10)
```

```python
# Quito esos encabezados de according...
f_codes = f_codes[~f_codes['Código'].str.contains('According to closing date of the study', na=False)]

f_codes.head(10)
```

```python
# Guardo en un nuevo archivo
output_path = r"C:\Users\Drakharys\Desktop\JustLearning\MUSIC\subject-info_codigos.csv"
f_codes.to_csv(output_path, index=False, sep=';', encoding='utf-8')

print(f"Archivo guardado en: {output_path}")
```
